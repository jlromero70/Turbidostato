```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
def indexes(y,thresh,peakdist):
    dy = np.diff(y)
    ind = []
    for i in range(len(dy)):
        if len(ind) > 0:
            if dy[i] < dy[int(i-1)] and np.abs(dy[i]) > thresh and i-peakdist>ind[-1]:
                ind.append(i)
        elif len(ind) == 0: 
            if dy[i] < dy[int(i-1)] and np.abs(dy[i]) > thresh:
                ind.append(i)
    return ind  
```

```python
df = pd.read_csv('fitvspics.csv', delimiter=';', decimal=',',header= None)
df.columns=['Tiempo','Plato1','Plato2','Plato3','Plato4','Plato5','Plato6','Plato7','Plato8']
tiempo = []
for med in df['Tiempo']:
    tiempo.append(med*8)
df['TiempoSeg'] = tiempo
del df['Tiempo']
df.head(n=9)
```

```python
thresh = 0.28
peakdist = 150
ind1 = indexes(df["Plato1"],thresh, peakdist)
tind1 = []
for i in ind1:
    tind1.append(df['TiempoSeg'][i])
fig, ax = plt.subplots(1, 1, figsize=(20, 5),sharex=True)
ax.plot(df["TiempoSeg"].tolist(),df["Plato1"].tolist(),c = 'r')#,ls = "--")
ax.scatter(tind1,[np.max(df["Plato1"])]*len(tind1),c = 'g',s = 20)
```
